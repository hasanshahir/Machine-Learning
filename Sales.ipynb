```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'TV_Spend': [230.1, 44.5, 17.2, 151.5, 180.8, 8.7, 57.5, 120.2, 8.6, 199.8],
    'Radio_Spend': [37.8, 39.3, 45.9, 41.3, 10.8, 48.9, 32.8, 19.6, 2.1, 2.6],
    'Newspaper_Spend': [69.2, 45.1, 69.3, 58.5, 58.4, 75.0, 23.5, 11.6, 1.0, 21.2],
    'Sales': [22.1, 10.4, 9.3, 18.5, 12.9, 7.2, 11.8, 13.2, 4.8, 10.6]
}
df = pd.DataFrame(data)
```

```python
X = df[['TV_Spend', 'Radio_Spend', 'Newspaper_Spend']]
Y = df['Sales']
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_Pred = model.predict(X)
```

```python
Min = Y.min()
Max = Y.max()

plt.scatter(Y,Y_Pred, color = "blue", label = "Actual vs Predicted")
plt.plot([Min, Max], [Min, Max], color = "red", linewidth = 2, label = "Best Fit Line")

plt.xlabel("Actual Price")
plt.ylabel('Predicted Price')
plt.title("Sales Price Prediction")
plt.legend()
plt.show()
```
