```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Year' : [2010.0,2012.0,2014.0,2016.0,2018.0,2019.0,2020.0,2021.0,2022.0,2023.0],
    'Mileage(km)' : [8000.0,60000.0,50000.0,40000.0,30000.0,20000.0,15000.0,10000.0,5000.0,2000.0],
    'EngineSize(L)' : [1.6,1.8,2.0,2.0,2.2,2.4,2.5,2.5,3.0,3.0],
    'Price(USD)' : [7000.0,9000.0,12000.0,15000.0,18000.0,21000.0,25000.0,27000.0,32000.0,35000.0]
}
df = pd.DataFrame(data)
```

```python
X = df[['Year', 'Mileage(km)', 'EngineSize(L)']]
Y = df['Price(USD)']
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_Pred = model.predict(X)
```

```python
# 1. Find the lowest and highest actual prices
min_price = Y.min()
max_price = Y.max()

# 2. Create the plot
plt.scatter(Y, Y_Pred, color='blue', label='Actual vs Predicted')

# 3. Draw the "Perfect" line (from min to max)
# This draws a line from (min, min) to (max, max)
plt.plot([min_price, max_price], [min_price, max_price], color='red', linewidth=2, label='Perfect Fit')

plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Car Price Prediction: Actual vs. Predicted')
plt.legend()
plt.show()
```
