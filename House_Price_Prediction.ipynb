```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Area' : [850,1000,1200,1500,1800,2000,2200,2500,3000,3500],
    'Price' : [75000, 95000, 110000, 150000,175000,200000,240000,270000,300000,350000]
}
df = pd.DataFrame(data)
```

```python
X = df[['Area']]
Y = df['Price']
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, color='blue', label='Actual Data') # The dots
plt.plot(X, Y_pred, color='red', linewidth=2, label='Best Fit Line') # The line
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('House Price Prediction: Linear Regression')
plt.legend()
plt.show()
```
