```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'Study_Hours': [2, 5, 8, 10, 4, 3, 9, 7, 6, 1],
    'Sleep_Hours': [8, 7, 6, 6, 8, 9, 6, 7, 8, 9],
    'Prev_Score': [70, 75, 85, 90, 72, 68, 88, 80, 78, 60],
    'Final_Grade': [72, 78, 88, 92, 75, 70, 91, 82, 79, 64]
}
df_student = pd.DataFrame(data)
```

```python
X = df_student[['Study_Hours', 'Sleep_Hours', 'Prev_Score']]
Y = df_student['Final_Grade']
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_Pred = model.predict(X)
```

```python
y_min = Y.min()
y_max = Y.max()

plt.scatter(Y, Y_Pred, color = "blue", label = "Actual vs Predicted")
plt.plot([y_min, y_max],[y_min, y_max], color = "red", linewidth = 2, label = "Best Fit Line")

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Student Grade Prediction")
plt.legend()
plt.show()
```
