```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 1. Prepare the Data (df now contains a table)
data = {
    'YearsExperience': [1.1, 1.3, 1.5, 2.0, 2.2, 3.0, 3.2, 3.7, 4.0, 5.0, 6.0, 7.0, 8.0, 9.0, 10.0],
    'Salary': [39343.0, 46205.0, 37731.0, 43525.0, 39891.0, 56642.0, 60150.0, 54445.0, 64445.0, 57189.0, 81363.0, 93940.0, 91738.0, 98273.0, 122391.0]
}
df = pd.DataFrame(data)
```

```python
# Separate input (X) and target (Y)
X = df[['YearsExperience']] #X is features
Y = df['Salary'] #Y is Target
```

```python
# 2. Build and Train the Linear Regression Model
model = LinearRegression()
model.fit(X, Y)
```

```python
# Make predictions (to draw the line)
Y_pred = model.predict(X)
```

```python
# 3. Visualize the Results
plt.scatter(X, Y, color='blue', label='Actual Data') # The dots
plt.plot(X, Y_pred, color='red', linewidth=2, label='Best Fit Line') # The line
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.title('Salary Prediction: Linear Regression')
plt.legend()
plt.show()
```
